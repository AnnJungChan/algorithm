- 배열에서 주어진 수들을 M번 더해서 가장 큰 수를 만드는 법칙

```python
n, m, k= map(int,input().split())
data = list(map(int,input().split()))
```

<스케치>

배열 중에 가장 큰 수  * 3 반복

배열 중에 가장 큰 수 제거

두번째로 큰 수 1번 반복

배열중에 가장 큰수 3번 반복

```python
count = 1
result = 0

if data.count(max(data)) > 1 :
    result = max(data) * m
else :
    while count <= m :
        data.sort(reverse = True)
        for _ in range(k) :
            result += data[0]
            count += 1
        for _ in range(1) :
            result += data[1]
            count += 1
```

```python
print(result)
```

## 해설

- 문제를 해결하기 위해선 가장 큰 수와 두번째로 큰 수만 있으면 된다.
- 그러므로 가장 큰 수와 두번째로 큰 수를 빼고는 모조리 버릴 것
- 가장 큰 수를 K번 더하고 두번째로 큰 수를 한번 더하는 연산만 반복된다.
- 버릴 것은 과감하게 버리자!

```python
n, m ,k = map(int,input().split())
data = list(map(int,input().split()))
```

```python
data.sort()
first = data[n-1]
second = data[n-2]
```

우리는 첫번째로 큰 수와 두번 째로 큰 수 밖에 이용하지 않는다. \\
그러므로 위와같이 두개의 숫자만 사용한다.

```python
result = 0

while True : 
    # 무한 루프

    for i in range(k) :
        if m == 0 :
            break
            #m이 0이라면 반복문 탈출
        result += first
        m -= 1
        # 더해지는 숫자에서 1을 하나씩 빼기

    if m == 0:
        break
    result += second
    # 두 번째로 큰 수 한번 더하기
    m -= 1
    # 더해지는 숫자에서 하나 빼기

print(result)
```

- 가장 큰 수가 반복되는 횟수를 확인
- 두번째로 큰 수가 반복되는 횟수를 확인

- 첫번째로 큰 수 : 6
- 두번째로 큰 수 : 5

K = 3일 때, 첫번째로 큰 수는 최대 3번 까지 밖에 반복될 수 없음

이 때, 반복되는 수열의 원소의 갯수는 4개 \\
그렇다면 이를 이용하여 첫번째로 큰 수가 얼마나 반복되는지를 파악가능하다.

숫자가 더해지는 횟수가 9라고 하자(M = 9)

```python
n = 5
m = 9
k = 3


int(m / (k+1)) * k + (m % (k+1))
```

그렇다면 수열은 2번 반복될 것이고 원소 4개 짜리 수열이 반복된다고 했을 때, 나머지에 대해서도 모두 첫번째 값이 채워질 것이다. \\
그러므로, 첫번째로 큰 숫자가 더해지는 횟수는 7

두번째 숫자가 더해지는 횟수는 수열이 반복될 때 마다 한번 씩 더해지므로, 그냥 수열의 반복 횟수이다.

```python
int(m/(k+1))
```

## 코드구현

```python
n, m, k = map(int,input().split())
data = list(map(int,input().split()))
```

```python
data.sort()
first = data[n-1]
second = data[n-2]

count = int(m/(k+1)) * k
count += m % (k+1)

result = 0 
result += (count) * first
result += (m-count) * second

print(result)
```
