가장 높은 숫자가 쓰인 카드 한 장을 뽑음

1. 숫자 카드는 행렬의 형태로 주어진다
2. 먼저 뽑고자 하는 카드가 있는 "행"을 선택
3. 그 다음 선택된 행에 포함된 카드 중 "가장 숫자가 낮은 카드"를 뽑음(지워나간다는 개념)
4. 처음에 카드를 골라낼 행 선택시, 이후에 해당 행에서 가장 낮은 카드를 뽑는 것을 고려하여, 최종적으로 가장 높은 카드를 뽑을 수 있도록 전략을 세움

# 나의 풀이

```python
n, m = map(int,input().split())
```

```python
result = []

for i in range(n) :
    data = list(map(int,input().split()))
    result.append(min(data))

print(max(result))
```

# 해설

- 각 행마다 가장 작은 수를 찾은 후, 그 수 중 가장 큰 수를 찾음
- 각 행마다 min()을 걸어주고, 그 값들을 모두 저장하여 max()를 걸어주면 문제가 해결된다.
- 이중 반복문으로 구현할 수도 있다.

### min(), max() 사용

```python
n, m = map(int,input().split())
```

```python
result = 0

for i in range(n) :
    
    data = list(map(int,input().split()))

    min_value = min(data) # 각 행에서 작은 값만 추출

    result = max(result, min_value) # 직전 값과 비교해서 더 큰 수를 추출

print(result)
```

### 2중 반복문 구조 이용

```python
n, m = map(int,input().split())
```

```python
result = 0
for i in range(n) :
    data = map(int,input().split())

    min_value = 10001 # 각 숫자는 10000이하이므로 절대로 10001이 최소숫자가 될 수 없다. 즉, 초기값을 정할 때, 절대로 우리가 정한 수가 최소값으로 뽑히지 말도록 하려고 10001로 초기값을 설정한다.

# 각 줄 비교

    for j in data :
        min_value=min(min_value, j) # 각 줄에서 최소값을 선정

# 전체 비교
    result = max(result,min_value) # 선정된 최소값을 result에 저장하고 다시 반복문으로 돌아감

print(result)
```
