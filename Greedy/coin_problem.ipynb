N = 1260 일 때의 경우를 파이썬 코드로 구현하라.

- 가장 큰 화폐 단위부터 거슬러 주면 된다.(간단한 아이디어)
- %= : 왼쪽 변수를 오른쪽 변수로 나눈 후, 그 나머지를 왼쪽 변수에 할당
- n %= coin : n을 coin으로 나눈 나머지를 n에 할당

```python
n = 1260
count = 0

# 큰 단위 화폐부터 차례대로 확인
coin_type = [500,100,50,10]

for coin in coin_type :
    count += n // coin
    n %= coin
```

```python
print(count)
```
