- N - 1

이 때, 수행하는 최소 횟수를 구해야한다.

# 나의 생각

```python
n, k = map(int,input().split())
```

```python
count = 0

while n != 1 :
    
    if n % k == 0 :
        n = int(n / k)
        count += 1

    else :
        n = n-1
        count += 1

print(count)
```

# 해설

- 최대한 많이 나누기를 수행하면 문제가 풀린다.
- 나누기를 많이 하는 이유는, 1을 빼는 것보다 나누는 것이 1로 더 빠르게 수렴하기 때문이다.
- 그러므로 아래의 과정을 반복한다

1. N이 K의 배수가 될 때 까지 1씩 빼기
2. N을 K로 나누기

## 답안 1

```python
n, k = map(int, input().split())

result = 0

while n >= k : 
    # n이 k 이상이라면, k로 계속 나누기(배수가 될 가능성을 최대한 찾기)

    while n % k != 0 : 
        # n이 k로 나누어 떨어지지 않는다면
        n -= 1
        result += 1
        # 1식 빼고 count를 1씩 늘려준다.

    n //= k
    # 그 후, 다 뺐으면 k로 나눠주고 이를 count한다
    result += 1

while n > 1:
    # 마지막으로 k미만인 수에 대해서
    n -= 1
    # -1씩 모두 빼주고
    result += 1
    # count 시켜준다

print(result)
```

## 답안 2

- 무한 루프를 이용

```python
n, k = map(int,input().split())
result = 0

while True :

    target = (n//k) * k
    # 우선 몫을 구한다음, 몫의 갯수만큼 k를 곱해주고
    result += (n - target)
    # 그 값을 전체에서 제거하면 나머지만 남는다, 나머지의 갯수를 우선 result에 저장한다.
    # 이 때, 나머지가 없으면 result에 0이 더하지고, 반복문이 탈출될 것이다.
    n = target
    # 그 후, target을 n에 저장한다. 이 말은 나머지를 제외한 값들을 n에 저장시킨다는 것이다.

    if n < k :
        # 만약, n이 k보다 작으면 반복문을 탈출한다.
        break

        # 여기까지 우리는 나머지에 대한 정보만 저장했고, 이제 몫에 대한 정보를 저장해야 한다.

    n //= k
    result += 1
    # 몫을 나눌 때 마다 1씩 count한다.

result += (n-1)
print(result)
```
